```python
import pandas as pd
import sqlite3
import yfinance as yf
```

```python
# create a sql connection and create the database stock
connection = sqlite3.connect('stock.db')
c = connection.cursor()
```

On crée une table pour les prix 

```python
# create price table with the primary tuple keys ( date, ticker) and the columns, date, ticker, price
query1 =('''CREATE TABLE IF NOT EXISTS price
              ([date] DATE, [ticker] VARCHAR(255), price INT, PRIMARY KEY(date,ticker))''')
c.execute(query1.replace('\n',' '))
```

```python
# create volume table with the primary tuple keys ( date, ticker) and the columns, date, ticker, volume 
query2 = """CREATE TABLE IF NOT EXISTS volume (
Date TEXT NOT NULL,
ticker TEXT NOT NULL,
volume REAL,
PRIMARY KEY(Date, ticker)
)"""
c.execute(query2.replace('\n',' '))
```

```python
# backtest inputs
bt_inputs = {'tickers': ['BA', 'UNH', 'MCD', 'HD'], #BA: Boeing, UNH: UnitedHealth, MCD: Mac Donalds, HD : HomeDepot
'start_date': '2019-01-01',
'end_date': '2021-06-01'}
```

```python
def download(bt_inputs, proxy = None):
    data = yf.download(tickers= bt_inputs['tickers'],
                       start = bt_inputs['start_date'],   
                       end = bt_inputs['end_date'],
                       interval = '1d',
                       prepost = True,
                       threads = True,
                       proxy = proxy)
    return data
```

```python
test = download(bt_inputs)
```

```python
#convert wide to long
volume_long = pd.melt(volume.reset_index(), id_vars='Date', value_vars=bt_inputs['tickers'], 
                      var_name = "ticker", value_name = "volume")
volume_long.head()
```

```python
# insert a in sql table prices and volume
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        return conn
    except Error as e:
        print(e)

    return conn
adj_close_long.to_sql("price", create_connection('stock.db'), if_exists="replace", index=False)
volume_long.to_sql("volume", create_connection('stock.db'), if_exists="replace", index=False)
```

```python
# inputs
select_tickers = bt_inputs['tickers']
start_date = bt_inputs['start_date']
end_date = bt_inputs['end_date']
# construct query: select the prices withe tickers in select_tickers 
#between start_date and end_date
query = ( """ select * from price where ticker in ('"""+ "','".join(select_tickers) + """') and Date >= '"""+ start_date + """' and Date < '""" + end_date + "'")
c.execute(query.replace('\n',' '))
#display the request
c.fetchall()
```

```python
#store the results of the request in a  dataframe result
c.execute(query.replace('\n',' '))
result = pd.DataFrame(c.fetchall(), columns =['Date','ticker','price'])
result
```
